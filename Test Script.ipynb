```python

# 1 second = 10^3 milliseconds
# 1 second = 10^6 microseconds
# 1 second = 10^9 nanoseconds

def unit_conv_time(unit_start, val, unit_end):
    
    mid_val = 0
    mid_unit = "s"
    
    if unit_start == "s":
        mid_val = val
    elif unit_start == "millis":
        mid_val = round(val / (10**3), 10)
    elif unit_start == "micros":
        mid_val = round((val / (10**6)), 10)
    elif unit_start == "nanos":
        mid_val = round(val / (10**9), 10)
    
    if unit_end == "s":
        return(mid_val)
    elif unit_end == "millis":
        return(mid_val * (10**3))
    elif unit_end == "micros":
        return(mid_val * (10**6))
    elif unit_end == "nanos":
        return(mid_val * (10**9))
```

```python
# 1 Byte = 8 bits
# 1 kB = 10^3 B
# 1 MB = 10^6 B
# 1 GB = 10^9 B

def unit_conv_bytes(unit_start, val, unit_end):
    
    mid_val = 0
    
    if unit_start[-1] == unit_end[-1]:
        if unit_start[0] == 'b' or unit_start[0] == 'B':
            mid_val = val
        elif unit_start[0] == 'k':
            mid_val = val * 10**3
        elif unit_start[0] == 'M':
            mid_val = val * 10**6
        elif unit_start[0] == 'G':
            mid_val = val * 10**9
    elif unit_start[-1] == 'B':
        if unit_start[0] == 'B':
            mid_val = 8 * val
        elif unit_start[0] == 'k':
            mid_val = 8 * val * 10**3
        elif unit_start[0] == 'M':
            mid_val = 8 * val * 10**6
        elif unit_start[0] == 'G':
            mid_val = 8 * val * 10**9
    elif unit_start[-1] == 'b':
        if unit_start[0] == 'b':
            mid_val = val / 8
        elif unit_start[0] == 'k':
            mid_val = (val * 10**3) / 8
        elif unit_start[0] == 'M':
            mid_val = (val * 10**6) / 8
        elif unit_start[0] == 'G':
            mid_val = (val * 10**9) / 8
    
    if unit_end[0] == 'B' or unit_end[0] == 'b':
        return(mid_val)
    
    if unit_end[0] == 'k':
        return(mid_val / 10**3)
    elif unit_end[0] == 'M':
        return(mid_val / 10**6)
    elif unit_end[0] == 'G':
        return(mid_val / 10**9)
```

```python
def unit_conv_transmission_rate(unit_num_start, val, unit_denom_start):
    if(unit_num_start == "b" and unit_denom_start == "s"):
        return val
    mid_denom = unit_conv_time(unit_denom_start, 1, "s")    #denominator is now seconds      value is in __/seconds
    mid_num = unit_conv_bytes(unit_num_start, val, "b")
    
    return mid_num/mid_denom
```

```python
#Speed of light in m/s

c = 299792458

class Node:
    def __init__(self, transmission_rate, unit_num, unit_denom):    #e.g.     150 bits / second => (150, "b", "s")
        self.transmission_rate = unit_conv_transmission_rate(unit_num, transmission_rate, unit_denom) #stores it in bits / second
        self.unit_num = "b"
        self.unit_denom = "s"
    
    def getTransmissionTime(self, packet):
        return packet.size / self.transmission_rate

class Link:
    def __init__(self, length, unit_length, speed, unit_speed):   #e.g. 120 metres in 120 metres / second => (120, 120, "m", s")
        self.length = length
        self.speed = speed
        self.unit_length = unit_length
        self.unit_speed = unit_speed
        self.propagation_delay = self.length / self.speed
    
    def getPropagationDelay(self):
        return self.length / self.speed
        
class Packet:
    def __init__(self, size, unit):
        self.size = unit_conv_bytes(unit, size, "b")
        self.unit = "b"
        
```

```python
class Network:
    
    
    def __init__(self, *args, node_count=None, link_count=None):
        
        self.node_count = node_count
        self.link_count = link_count
        self.network_ds = args
    
    def passTwoPackets(self, packet1, packet2):           #3 Nodes
        
        total_time = 0
        
        print("Transmission time for packet 1 of size " + str(packet1.size) + " b")
        print("Going through Link 1 of transmission rate " + str(self.network_ds[0].transmission_rate) + " bits/seconds")
        print(str(self.network_ds[0].getTransmissionTime(packet1)) + " seconds" + "\n============================\n")
        print("Transmission time for packet 2 of size " + str(packet2.size) + " b")
        print("Going through link 1 of transmission rate " + str(self.network_ds[0].transmission_rate) + " bits/seconds")
        print(str(self.network_ds[0].getTransmissionTime(packet2)) +  " seconds" + "\n============================\n")
        print("Propagation delay for link 1\n")
        print(str(self.network_ds[1].getPropagationDelay()) + " seconds")
        print("============================\n")
        print("Queueing Delay")
        print("8D1/T2 = " + str(self.network_ds[2].getTransmissionTime(packet1)))
        print("8D2/T1 = " + str(self.network_ds[0].getTransmissionTime(packet2)))
        queuing_delay = max(0, self.network_ds[2].getTransmissionTime(packet1) 
                            - self.network_ds[0].getTransmissionTime(packet2))
        
        print(str(queuing_delay) + " seconds")
        
        if queuing_delay == 0.0:
            print("there is no queueing delay")
        
        else:
            max_size = self.network_ds[0].getTransmissionTime(packet2) * self.network_ds[2].transmission_rate
            min_size = self.network_ds[2].getTransmissionTime(packet1) * self.network_ds[0].transmission_rate
            print("For queuing delay to be 0")
            print("Packet 1 can be of maximum size " + str(max_size) + " bits")
            print("Packet 2 can be of minimum size " + str(min_size) + " bits")
        
        print("============================")
        print("Transmission time for packet 2 of size " + str(packet2.size) + " b")
        print("Going through link 2 of transmission rate " + str(self.network_ds[2].transmission_rate) + "bits/seconds");
        print("Transmission time for packet 2 going through link 2 " + str(self.network_ds[2].getTransmissionTime(packet2)) + " seconds \n===================\n")
        print("Propagation delay for link 2 " + str(self.network_ds[3].getPropagationDelay()) + " seconds")
        total_time = self.network_ds[0].getTransmissionTime(packet1) + self.network_ds[0].getTransmissionTime(packet2) + self.network_ds[1].getPropagationDelay() + queuing_delay + self.network_ds[2].getTransmissionTime(packet2) + self.network_ds[3].getPropagationDelay()
        
        print("Total time")
        print(str(total_time) + " seconds")
    
    def rtt_2_nodes(self, packet1):                           #2 nodes
        print("Transmission time for packet 1 of size " + str(packet1.size) + " b")
        print("Going through link 1 of transmission rate " + str(self.network_ds[0].transmission_rate) + " bits/seconds")
        print(str(self.network_ds[0].getTransmissionTime(packet1)) + " seconds" + "\n=========================\n")
        print("Propagation delay for link 1\n")
        print(str(self.network_ds[1].getPropagationDelay()) + " seconds")
        print("============================\n")
        total_time = self.network_ds[0].getTransmissionTime(packet1) + self.network_ds[1].getPropagationDelay()
        print("One way time " + str(total_time) + " seconds")
        return 2 * total_time
    
```

```python
#Aggregate of Peaks
#Peaks of Aggregates

class stats:
    
    def __init__(self, *args):
        self.flows = args
        self.lengths = []
        for i in self.flows:
            self.lengths.append(len(i))
    
    def peak_rates(self):
        for i in self.flows:
            print("Flow peak   " + str(max(i)))
    
    def sum_of_peaks(self):
        peaks_arr = []
        for i in self.flows:
            peaks_arr.append(max(i))
        
        return sum(peaks_arr)
    
    def peak_of_agg(self):
        agg_arr = []
        
        for i in range(len(self.flows[0])):
            agg_single = 0
            for j in range(len(self.flows)):
                agg_single = agg_single + self.flows[j][i]
            agg_arr.append(agg_single)
        return(max(agg_arr))
```

```python
#Circuit Switching vs Packet Switching

#We are provided circuit setup time, hops, propagation delay per hop, transmission rate at each node

def circuit_switching(circuit_setup_time=0, circuit_setup_unit = "",
                     hop_count = 0,
                     prop_delay_per_hop = 0, prop_delay_unit = "",
                     trans_rate_each_node = 0, trans_rate_unit_num = "", trans_rate_unit_denom = "",
                     packet_size = 0, unit_start = 0):
    total_time = 0
    pkt = unit_conv_bytes(unit_start, packet_size, "b")
    circuit_setup = unit_conv_time(circuit_setup_unit, circuit_setup_time, "s")
    prop_del = unit_conv_time(prop_delay_unit, prop_delay_per_hop, "s")
    trans_rate = unit_conv_transmission_rate(trans_rate_unit_num, trans_rate_each_node, trans_rate_unit_denom)
    total_time = circuit_setup + (hop_count * prop_del) + (pkt / trans_rate)
    print("Total Time in circuit switching = s + (x/b) + kd = " + str(total_time))
    return total_time

#t = (x/b) + kd + (k-1)(p/b)

def packet_switching(circuit_setup_time = 0, circuit_setup_unit = "",
                    hop_count = 0,
                    prop_delay_per_hop = 0, prop_delay_unit = "",
                    trans_rate_each_node = 0, trans_rate_unit_num = "", trans_rate_unit_denom = "",
                    packet_size = 0, unit_start_packet = "",
                    msg_size = 0, unit_start_msg = ""):
    total_time = 0
    pkt = unit_conv_bytes(unit_start_packet, packet_size, "b")
    msg = unit_conv_bytes(unit_start_msg, msg_size, "b")
    prop_del = unit_conv_time(prop_delay_unit, prop_delay_per_hop, "s")
    trans_rate = unit_conv_transmission_rate(trans_rate_unit_num, trans_rate_each_node, trans_rate_unit_denom)
    
    #no circuit setup time
    total_time = (hop_count * prop_del) + (msg / trans_rate) + (hop_count - 1) * (pkt / trans_rate)
    print("Total time in packet switching = (x/b) + kd + (k-1)(p/b) = " + str(total_time))
```

```python
# INTERNET #

#Classless CIDR
class ip:
    def __init__(self, b1, b2, b3, b4):
        self.b1 = b1
        self.b2 = b2
        self.b3 = b3
        self.b4 = b4
    
    def passIP(self, ip_address):
        print("mask top then ip")
        print(bin(self.b1)[2:].zfill(8) + "." + bin(self.b2)[2:].zfill(8) + "." + bin(self.b3)[2:].zfill(8) + "." + bin(self.b4)[2:].zfill(8))
        print(bin(ip_address.b1)[2:].zfill(8) + "." + bin(ip_address.b2)[2:].zfill(8) + "." + bin(ip_address.b3)[2:].zfill(8) + "." + bin(ip_address.b4)[2:].zfill(8))
        print("---------------------------------------------------")
        print(bin(self.b1 & ip_address.b1)[2:].zfill(8) + "." + bin(self.b2 & ip_address.b2)[2:].zfill(8) + "." + bin(self.b3 & ip_address.b3)[2:].zfill(8) + "." + bin(self.b4 & ip_address.b4)[2:].zfill(8))
        print(str(self.b1 & ip_address.b1) + "." + str(self.b2 & ip_address.b2) + "." + str(self.b3 & ip_address.b3) + "." + str(self.b4 & ip_address.b4))
        
    def binary_rep(self):
        print(bin(self.b1)[2:].zfill(8) + "." + bin(self.b2)[2:].zfill(8) + "." + bin(self.b3)[2:].zfill(8) + "." + bin(self.b4)[2:].zfill(8))
        
    def net_bits_count(self):
        bit_rep = bin(self.b1)[2:].zfill(8) + " " + bin(self.b2)[2:].zfill(8) + " " + bin(self.b3)[2:].zfill(8) + " " + bin(self.b4)[2:].zfill(8)
        
        net_count = 0
        
        for i in bit_rep:
            if i == "1":
                net_count += 1
            elif i == "0":
                break
        
        return net_count
    
    def subnet_range(self, ip):
        net_count = self.net_bits_count()
        subnet_count = 32 - net_count - 8
        print("The subnet range is from " + str(ip.b1) + "." + str(ip.b2) + "." + "0" + "." + "1", end=" ")
        print("to " + str(ip.b1) + "." + str(ip.b2) + "." + str((2**subnet_count) - 1) + "." + "254")
    
    def host_count(self, ip):
        net_bit_count = self.net_bits_count()
        host_bit_count = 32 - self.net_bits_count()
        print("The number of hosts can be " + str((2**host_bit_count) - 2))
                
```

```python
import math

class tcp_udp_ip:
    def __init__(self,
                 rtt = 0, rtt_unit = "",
                 receiver_advertised_window = 0, receiver_advertised_window_unit = "",
                 threshold = 0, threshold_unit = "",
                 mss = 0, mss_unit= ""):
        self.rtt = unit_conv_time(rtt_unit, rtt, "millis")
        self.receiver_advertised_window = unit_conv_bytes(receiver_advertised_window_unit, receiver_advertised_window, "kB")
        receiver_advertised_window_unit = "kB"
        self.threshold = unit_conv_bytes(threshold_unit, threshold, "kB")
        self.mss = unit_conv_bytes(mss_unit, mss, "kB")
        self.mss_unit = "kB"
    
    def full_window_time(self):               #Time until full window sent
        cwnd = 2 * self.mss
        transmit = self.mss
        i = 1
        threshold_reached = False
        
        while transmit < self.receiver_advertised_window:
            print("Transmission number " + str(i) + " Transmission is " + str(transmit) + " kB, " + "Congestion window is " + str(cwnd) + " kB")
            print("Time passed is " + str((i * self.rtt)) + " milliseconds")
            i += 1
            
            transmit = cwnd
            
            if(cwnd * 2) >= self.threshold and not threshold_reached:
                cwnd = self.threshold
                threshold_reached = True
            
            if threshold_reached:
                cwnd += self.mss
            else:
                cwnd *= 2
            
    def n_timeout(self, timeout_n, transmissions_n):
        cwnd = 2 * self.mss
        transmit = self.mss
        i = 1
        threshold_reached = False
        
        while i < timeout_n:
            print("Transmission number " + str(i) + " Transmission is " + str(transmit) + " kB, " + "Congestion window is " + str(cwnd) + " kB")
            print("Time passed is " + str((i * self.rtt)) + " milliseconds")
            i += 1
            transmit = cwnd
            
            if (cwnd * 2) >= self.threshold and not threshold_reached:
                cwnd = self.threshold
                threshold_reached = True
                continue
                
            if threshold_reached:
                cwnd += self.mss
            else:
                cwnd *= 2
            
            if cwnd >= self.receiver_advertised_window:
                cwnd = self.receiver_advertised_window
        
        temp_threshold = int(transmit / 2)
        
        print("Transmission number " + str(i) + " Transmission is " + str(transmit) + " kB - timeout - reset threshold to " + str(temp_threshold) + "kB")
        print("Time passed is " + str(i * self.rtt) + " milliseconds")
        
        i += 1
        cwnd = 2 * self.mss
        transmit = self.mss
        threshold_reached = False
        
        while i <= transmissions_n:
            print("Transmission number " + str(i) + " Transmission is " + str(transmit) + " kB, " + "Congestion window is " + str(cwnd) + " kB")
            print("Time passed is " + str((i * self.rtt)) + " milliseconds")
            i += 1
            transmit = cwnd
        
            if (cwnd * 2) >= temp_threshold and not threshold_reached:
                cwnd = temp_threshold
                threshold_reached = True
                continue

            if threshold_reached:
                cwnd += self.mss
            else:
                cwnd *= 2

            if cwnd >= self.receiver_advertised_window:
                cwnd = self.receiver_advertised_window
    
    def n_transmissions(self, n):
        cwnd = 2 * self.mss
        transmit = self.mss
        i = 1
        threshold_reached = False
        
        while i <= n:
            print("Transmission number " + str(i) + " Transmission is " + str(transmit) + " kB, " + "Congestion window is " + str(cwnd) + " kB")
            print("Time passed is " + str((i * self.rtt)) + " milliseconds")
            i += 1
            transmit = cwnd
            
            if(cwnd * 2) >= self.threshold and not threshold_reached:
                cwnd = self.threshold
                threshold_reached = True
                continue
            
            if threshold_reached:
                cwnd += self.mss
            else:
                cwnd *= 2
            
            if cwnd >= self.receiver_advertised_window:
                cwnd = self.receiver_advertised_window
    
    
```

```python
import math

def bit_err_prob(peb = 0,
                 p = 0,
                 L = 0, unit_L = "",
                 hop_count = 0,
                 calc_val = ""):
    if(calc_val == ""):
        print('no input provided !')
    elif(calc_val == "p"):
        p = 1 - (1 - peb)**(unit_conv_bytes(unit_L, L, "b")*hop_count)
        print("Probability of error = ", p)
    elif(calc_val == "L"):
        l = math.log(1-p)/(hop_count * math.log(1-peb))
        print("Frame size in bits = ", l)
    elif(calc_val == "hop_count"):
        hop_count = math.log(1-p)/(unit_conv_bytes(unit_L, L, "b") * math.log(1 - peb))
        print("hop count = ", hop_count)
    elif(calc_val == "peb"):
        peb = 1 - ((1 - p)**(1/(unit_conv_bytes(unit_L, L, "b") * hop_count)))
        print("Bit error rate = ", peb)
    else:
        print("incorrect input provided !")
```

```python

def showpoly(a):
    str1 = ""
    nobits = len(a)

    for x in range(nobits - 2):
        if a[x] == '1':
            if len(str1) == 0:
                str1 += f'x^{nobits - x - 1}'
            else:
                str1 += f' + x^{nobits - x - 1}'

    if a[nobits - 2] == '1':
        if len(str1) == 0:
            str1 += 'x'
        else:
            str1 += ' + x'

    if a[nobits - 1] == '1':
        str1 += ' + 1'

    print(str1)

def toList(x):
    l = []
    for i in range(len(x)):
        l.append(int(x[i]))
    return l

def toString(x):
    str1 = ""
    for i in range(len(x)):
        str1 += str(x[i])
    return str1

def divide(val1, val2):
    print(val1)
    a = toList(val1)
    b = toList(val2)
    working = toString(val1) + "\n"
    res = ""
    addspace = ""

    while len(b) <= len(a) and a:
        if a[0] == 1:
            del a[0]
            for j in range(len(b) - 1):
                a[j] ^= b[j + 1]
            if len(a) > 0:
                working += addspace + toString(b) + "\n"
                working += addspace + "-" * len(b) + "\n"
                addspace += " "
                working += addspace + toString(a) + "\n"
                res += "1"
        else:
            del a[0]
            working += addspace + "0" * len(b) + "\n"
            working += addspace + "-" * len(b) + "\n"
            addspace += " "
            working += addspace + toString(a) + "\n"
            res += "0"

    print("Result is\t", res)
    print("Remainder is\t", toString(a))

    print("Working is\t\n\n", res.rjust(len(val1)), "\n")
    print("-" * len(val1), "\n", working)

    return toString(a)
```

```python
class stop_and_wait_network:
    
    def __init__(self,
                 sender,
                 link,
                 receiver):
        self.sender = sender
        self.receiver = receiver
        self.link = link
    
    def getTF_sender(self, frame):
        return self.sender.getTF(frame)
    
    def getTF_receiver(self, frame):
        return self.receiver.getTF(frame)
    
    def getPropDelay(self):
        return self.link.getPropDelay()
    
    def getTS(self, frame_send, frame_return):
        tf_sender = self.getTF_sender(frame)
        tf_receiver = self.getTF_receiver(frame_return)
        
        processing_delay_sender = self.sender.proc_time
        processing_delay_receiver = self.receiver.proc_time
        
        propagation_delay = self.getPropDelay()
        
        return tf_sender + propagation_delay + processing_delay_receiver + tf_receiver + propagation_delay + processing_delay_sender
    
    def getA_sender(self, frame_send, frame_return):
        ts = self.getTS(frame_send, frame_return)
        tf = self.getTF_receiver(frame_send)
        
        return ts/tf
    
    def prob_err_free_delivery(self):          #probability of Ts
        return (1 - self.sender.prob_frame_error) * (1 - self.receiver.prob_frame_error)
    
    def prob_err_delivery(self):               #probability of TFail
        return 1 - self.prob_err_free_delivery()
    
    def avg_frame_transmissions_for_success(self):        #E
        return 1/self.prob_err_free_delivery()
    
    def avg_packet_delay_in_arq(self, frame):               #D
        E = self.avg_frame_transmissions_for_success()
        Tfail = self.prob_err_delivery()
        Ts = self.getTS(frame, frame)
        return (E - 1) * Tfail + Ts
    
    def avg_packet_throughput(self, frame):
        return 1/self.avg_packet_delay_in_arq(frame)
    
    def efficiency(self, frame):
        return self.getTF_sender(frame) / self.avg_packet_delay_in_arq(frame)
    
    def opt_timeout_symmetrical_case(self, prob, frame):
        Tfail = 1 - (1 - prob)**2
        print("E = 1/(1-p)^2 = ", 1/((1-prob)**2))
        E = 1/(1-prob)**2
        Ts = self.getTS(frame, frame)
        print("D = E * Ts ", (1/((1-prob)**2) - 1) * Tfail + Ts)
        D = E * Ts
        print('efficiency = Tf / D = ', self.getTF_sender(frame) / D)
        
    def opt_timeout_no_error_ack(self, prob, frame):
        E = 1/(1 - prob)
        Ts = self.getTS(frame, frame)
        D = Ts / (1 - prob)
        print("E = 1/(1-p) = ", E)
        print("D = Ts / (1 - prob) = ", D)
        print("efficiency = Tf / D = ", self.getTF_sender(frame) / D)
    
    def symmetric_case(self, prob, frame):
        Tfail = self.prob_err_delivery()
        Ts = self.getTS(frame, frame)
        print("D = ( 1 / ( 1  p )^2  1 ) * TFail + Ts", (1/(1-prob)**2 - 1)*Tfail + Ts)
        print("E = 1/(1-p)^2 = ", 1/(1 - prob)**2)
    
    def no_error_ack(self, prob, frame):
        Tfail = self.prob_err_delivery()
        Ts = self.getTS(frame, frame)
        print("D = p/(1-p) * Tfail + Ts = ", (prob / (1 - prob)) * Tfail + Ts)
        print("E = 1/1(1-p) = ", 1/(1 - prob))
    
    def proc_time_zero_case(self, prob, frame):
        print("Proc = 0, q = 0, r = p, E = 1/(1-p), timeout = Ta + 2*(proc + prop)")
        print("Tfail = Ts, D = E * Ts, efficiency = (1-p)Tf/Ts")
    
```

```python
def broadcast_time_wasted(p = 0, n = 0):
    p_collision = 1 - (1-p)**n - n*(p*(1-p)**(n-1))
    p_succesful = n*(p*(1-p)**(n-1))
    p_empty = (1-p)**n
    print("p_collision = ", p_collision, " (time wasted due to collisions)")
    print("p_succesful = ", p_succesful, "(time spent being succesful)")
    print("p_empty = ", p_empty, "(time wasted due to link being empty)")
```

```python
broadcast_time_wasted(p = 0.01, n = 10)
```
